```python
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go

init_notebook_mode(connected=True)

import pandas as pd

###############################################

# Modules for widgets
import ipywidgets as wg
from IPython.display import display

# Modules for delay
import time
from collections import Counter

df_economicfreedom = pd.read_csv('efw_cc.csv', encoding = "ISO-8859-1", low_memory=False)
df_terrorism = pd.read_csv('globalterrorismdb_0718dist.csv', encoding = "ISO-8859-1", low_memory=False)
df_wgi = pd.read_csv('3a2d6bc1-8ed7-4c33-8db5-b001259e1d3a_Data.csv', encoding = "ISO-8859-1", low_memory=False)

df_economicfreedom.fillna(0, inplace=True)
df_terrorism.fillna(0, inplace=True)
df_wgi.fillna(0, inplace=True)
```

```python
western_countries = ['Greece', 'United States', 'Italy', 'East Germany (GDR)', 'West Germany (FRG)', 'Switzerland', 'Spain', 'Ireland',
                     'United Kingdom', 'Netherlands', 'Belgium', 'Canada', 'Australia', 'Sweden', 'Poland', 'Austria', 'Czechoslovakia',
                    'France', 'Portugal', 'Cyprus', 'New Zealand', 'Andorra', 'Malta', 'Japan' ,'Denmark', 'Norway', 'Vatican City', 'Luxembourg',
                     'Iceland', 'Finland', 'Hungary','Germany', 'Lithuania', 'Slovenia', 'Czech Republic', 'Latvia', 'Estonia', 'Slovak Republic',
                    'International']
```

```python
year = wg.IntSlider(
    value=2007,
    min=1970,
    max=2016,
    step=1,
    description='Year: ',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
```

```python
def donut_maker2(year):
    dt = []
    cyear = df_economicfreedom.loc[df_economicfreedom['year'] == year]
    tyear = df_terrorism.loc[df_terrorism['iyear'] == year]
    countries = tyear['country_txt'].unique()

    for country in countries:
        cur_ct = cyear[cyear['countries'] == country]
        cur_ct_terrorism = tyear[tyear['country_txt'] == country]
        type_economic_freedom = cur_ct['4d_control_movement_capital_ppl']
        aanslagen = 0
        for row in cur_ct_terrorism['country_txt']:
            aanslagen += 1
        if country in western_countries:
            trace0 = go.Scatter(
                x = type_economic_freedom,
                y = [aanslagen],
                mode = 'markers',
                marker = dict(
                    size = 10,
                    color = 'rgb(0, 256, 0)'
                ),
                name = country,
                text = country
            )
            dt.insert(0, trace0)
        else:
            trace0 = go.Scatter(
                x = type_economic_freedom,
                y = [aanslagen],
                mode = 'markers',
                marker = dict(
                    size = 10,
                    color = 'rgb(256, 0, 0)'
                ),
                name = country,
                text = country
            )
            dt.append(trace0)

    layout = go.Layout(
        autosize = True,
        title = go.layout.Title(
            text ="Aantal aanslagen tegenover economische vrijheid (" + str(year) + ")",
        ),
        xaxis=go.layout.XAxis(
            type = 'log',
            title = 'Judicial Independence',
            autorange = True,
        ),
        yaxis = go.layout.YAxis(
            type = 'log',
            title = 'Aanslagen',
            autorange = True,
        ),
    )
    if dt == []:
        print("No data available for this time period.")
    fig = go.Figure(data= dt, layout=layout)
    iplot(fig)
```

```python
wg.interact(donut_maker2, year = year)
```
